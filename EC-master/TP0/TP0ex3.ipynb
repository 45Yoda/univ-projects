<b> Exercício 3 - Corpos finitos primos </b>

Na primeira parte deste exercício tivemos de criar quatro corpos finitos sendo eles P=37,163,263,1009.Para isso usamos a funçao FiniteField().

```python
p1 = FiniteField(37)
print(p1)

p2 = FiniteField(163)
print(p2)

p3 = FiniteField(263)
print(p3)

p4 = FiniteField(1009)
print(p4)
```

Uma vez criados os corpos, foi criado um gráfico para cada um destes através da função plot do SageMath, como podemos ver:

```python
import matplotlib.pyplot as plt
graph = plt.plot([x**(37-2) for x in p1])
plt.show(graph)
```

```python
graph = plt.plot([x**(163-2) for x in p2])
plt.show(graph)
```

```python
graph = plt.plot([x**(263-2) for x in p3])
plt.show(graph)
```

```python
graph = plt.plot([x**(1009-2) for x in p4])
plt.show(graph)
```

Por último, determina-se um elemento primitivo para cada um dos corpos finitos primos, verificando-se depois a proposição exposta no enunciado para cada um desses elementos, da qual se mostra sempre verdadeira.

```python
ep1 = p1.primitive_element()
print(ep1)
n = mod(0,37-1)
ep1^n == 1
```

```python
ep2 = p2.primitive_element()
print(ep2)
n = mod(0,163-1)
ep2^n == 1
```

```python
ep3 = p3.primitive_element()
print(ep3)
n = mod(0,263-1)
ep3^n == 1
```

```python
ep4 = p4.primitive_element()
print(ep4)
n = mod(0,1009-1)
ep4^n == 1
```
